#### Latitude vs. Temperature Plot

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
weather_csv = "Charts/weather_data.csv"
weather_df = pd.read_csv(weather_csv)
weather_df.head()
```

#### Latitude vs. Humidity Plot

```python
lat_list = weather_df['Lat']
x_axis = np.arange(-90, 90, 30)
humid_list = weather_df['Humidity']

plt.scatter(lat_list, humid_list, marker="o", facecolors="lightblue", 
            edgecolors="navy", alpha=.6)
# plt.xlim(-80, 80)
plt.ylim(0, 110)

plt.grid()

plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Humidity by Latitude", weight="bold")

plt.show()
plt.savefig('Charts/Humidity.png')
```

#### Latitude vs. Cloudiness Plot

```python
lat_list = weather_df['Lat']
x_axis = np.arange(-90, 90, 30)
cloud_list = weather_df['Cloudiness']
# opacity = [int(i) for i in cloud_list]
date_max = weather_df['Date'].max()
date_min = weather_df['Date'].min() #CONVERT DATES


plt.scatter(lat_list, cloud_list, marker="o", facecolors="lightblue", 
            edgecolors="navy", alpha=.6)

# plt.ylim(0, 120)

plt.grid()

plt.xlabel("Latitude")
plt.ylabel("Cloudiness")
plt.title(f"City Latitude vs. Cloudiness ({date_min}-{date_max})", weight="bold")

plt.show()
plt.savefig('Charts/Cloudiness.png')
```

#### Latitude vs. Wind Speed Plot

```python
lat_list = weather_df['Lat']
x_axis = np.arange(-90, 90, 30)
wind_list = weather_df['Wind Speed']

plt.scatter(lat_list, wind_list, marker="o", facecolors="lightblue", 
            edgecolors="navy", alpha=.6)
# plt.xlim(-80, 80)
# plt.ylim(0, 120)

plt.grid()

plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title(f"City Latitude vs. Wind Speed ({date_min}-{date_max})", weight="bold")

plt.show()
plt.savefig('WindSpeed.png')
```

```python
lat_list = weather_df['Lat']
x_axis = np.arange(-90, 90, 30)
humid_list = weather_df['Humidity']
temp_size = weather_df['Max Temp']

plt.scatter(lat_list, humid_list, marker="o", facecolors="lightblue", 
            edgecolors="blue", alpha=.5, s=(temp_size-200)*5, label="Max Temp")
# plt.xlim(-80, 80)
plt.ylim(0, 120)

plt.grid()

plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Humidity by Latitude", weight="bold")

plt.show()
```
